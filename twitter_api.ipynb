Twitter API to search tweets

```python
import tweepy
import pandas as pd
```

Step 2. Authenticate using OAth method

```python
API_key = '-------'
API_secret = '---------'
bearer_token= '----------'
access_secret = '----------'
```

```python
auth = tweepy.OAuthHandler(API_key, API_secret)
```

```python
auth.set_access_token(bearer_token, access_secret)
```

```python
api = tweepy.API(auth)
```

Step 3. Begin your API requests!


Let's search for 50 tweets, containing the hashtag #covid19, in English, and posted Feburary7th,2022.

```python
for tweet in tweepy.Cursor(api.search_tweets,q="#covid19",
                        lang="en",
                        since_id="2022-02-07").items(50):
    print(tweet)

```

Step 3. Create a dataframe with the attributes of interest

```python
tweets = []
date = []
tweet_source = []
likes = []
```

```python
for tweet in tweepy.Cursor(api.search_tweets,q="#covid19",
                        lang="en",
                        since_id="2022-02-07").items(50):
    tweets.append(tweet.text)
    date.append(tweet.created_at)
    likes.append(tweet.favorite_count)
    tweet_source.append(twet.source)
```

```python
twitter_df = pd.DataFrame({'Tweets':tweets, 'Date_posted':date, "favourite":likes, "Device":tweet_source})
```
